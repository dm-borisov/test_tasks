**Рекламная кампания стартовала вчера с дневным бюджетом 40$. Половина была израсходована к полудню, а 80% оставшегося бюджета было потрачено между полуднем и временем закрытия. Сколько долларов не было потрачено?**

Решение:
1.	К полудню осталась только половина бюджета рекламной кампании: $0.5 × 40 = 20\$$
2.	80% от оставшегося бюджета было потрачено между полуднем и временем закрытия: $0.8 × 20 = 16\$$
3.	В итоге осталось: $40 – 20 – 16 = 4\$$

Ответ: не было потрачено 4 доллара.

**5 идентичных рекламных кампаний работали 24 дня по 6 часов в день, потрачено было 120 долларов. Сколько дней они работали бы на 216 долларов, если бы 9 одинаковых кампаний работали бы по 8 часов в день?**

Для начала введем обозначения:
*   c1, c2 – количество кампаний в первом и во втором случае
*   d1, d2 – сколько дней проработали кампании
*   h1, h2 – сколько часов кампании работали в день
*   e1, e2 – затраты на кампании
*   r – сколько кампания тратит за один час работы

Решение:
1.	Так как кампании идентичны, то они тратят одинаковое количество за один час работы: $r=\dfrac{s}{c × d × h}$
2.	В обоих случаях траты за час будут одинаковы, поэтому мы можем записать: $r=\dfrac{s1}{c1 × d1 × h1}=\dfrac{s2}{c2 × d2 × h2}$
3.	Из равенства выше выведем d2: $d2 = \dfrac{s1 × c1 × d1 × h1}{s2 × c2 × h2}=\dfrac{120 × 5 × 24 × 6}{216 × 9 × 8} = 18$

Ответ: 18 дней.


**Дизайнеры создали 200 рекламных баннеров для двух рекламных кампаний. 80 из них не использовались ни в кампании №1, ни в кампании №2, 60 использовались только в кампании №1. И для каждого рекламного баннера, который использовался в обеих кампаниях, приходится 3 баннера, которые использовались только в кампании №2. Сколько баннеров было использовано в обеих рекламных кампаниях?**

Решение:
1.	Для начала найдем, сколько баннеров использовали обе кампании: $200 – 80 = 120$
2.	Обозначим за s количество баннеров, которые использую обе кампании.
3.	Так как, на каждый баннер, используемый обеими кампаниями, приходится 3 баннера, которые использовались только в кампании 2, то последнее число будет равно 3s
4.	Количество использованных баннеров можно выразить следующим способом: 60 + $3s = 120$ (В 60 входят и баннеры, используемые кампанией 1, и баннеры, которые используются обеими кампаниями)
5.	Из выражения выше, найдем s: $s = \dfrac{120 – 60}{3} = 20$

Ответ: 20 баннеров было использовано в обеих рекламных кампаниях.

**Средний итоговый балл стажера по 4 модулям составляет 78 баллов. Сколько баллов должен получить стажер за 5-й модуль, чтобы средний балл по всему заданию составил 80?**
Решение:
1.	Пусть xi – номер модуля.
2.	Тогда среднее за 4 модуля можно выразить следующим равенством: $\dfrac{x1+x2+x3+x4}{4}=78$
3.	Среднее за 5 модулей выразим так: $\dfrac{x1+x2+x3+x4+x5}{5}=80$
4.	Из шага 2 выразим $x1+x2+x3+x4$: $x1+x2+x3+x4 = 78 × 4 = 312$
5.	Подставим результат выше в шаг 3 и выразим из него x5: $x5 = 80 \times 5 – 312 = 88$

Ответ: чтобы средний балл по всему заданию составил 80, стажер должен набрать 88 баллов за модуль 5.


**Заказанные в приложении товары доставляются на автомобиле. Автомобиль проезжает 260 км со средней скоростью 80 км/ч. На обратном пути машина движется со средней скоростью 100 км/ч. Насколько быстрее был обратный путь? Ответ указать в минутах.**
Решение:
1.	Пусть s – путь, который проезжает автомобиль, v1, v2 – скорость в пути и на обратном пути.
2.	Уравнение пути выглядит следующим образом: $s = v × t$, где t – время в пути.
3.	Выразим t: $t = \dfrac{s}{v}$
4.	В первый раз, автомобиль потратил: $t1 = \dfrac{260}{80} = 3.25$ часа, или в минутах $3.25 × 60$ = 195 минут
5.	На обратном пути, автомобиль затратил: $t2 = \dfrac{260}{100}$ = 2.6 часа, или в минутах $2.6 × 60 = 156$ минут
6.	Найдем разницу между временами: $Δt = 195 – 156 = 39$ минут

Ответ: на обратный путь автомобиль затратил на 39 минут меньше.

# Задание
За каждую заключенную сделку менеджер получает бонус, который рассчитывается
следующим образом.
1) За новые сделки менеджер получает 7 % от суммы, при условии, что статус
оплаты «ОПЛАЧЕНО», а также имеется оригинал подписанного договора с
клиентом (в рассматриваемом месяце).

2) За текущие сделки менеджер получает 5 % от суммы, если она больше 10 тыс.,
и 3 % от суммы, если меньше. При этом статус оплаты может быть любым,
кроме «ПРОСРОЧЕНО», а также необходимо наличие оригинала подписанного
договора с клиентом (в рассматриваемом месяце).

Бонусы по сделкам, оригиналы для которых приходят позже рассматриваемого
месяца, считаются остатком на следующий период, который выплачивается по мере
прихода оригиналов. Вычислите остаток каждого из менеджеров на 01.07.2021.

Остаток каждого менеджера на 01.07.2021 - это остатока, рассчитанный для договоров, оригиналы которых пришли в июне. Создадим DataFrame, где соберем все майские договорые, оригиналые которых пришли в июне.

```python
june_from_may = (df["date"] == "2021-05") & (df["receiving_date"].dt.month == 6)
df_may_remainder = df[june_from_may].copy()
df_may_remainder.head()
```

Создадим функцию, которая будет считать бонусы в зависимости от условий задания

```python
def calculate_bonus(df):
  if df["new/current"] == "новая" and df["status"] == "ОПЛАЧЕНО":
    return 0.07 * df["sum"]
  elif df["new/current"] == "текущая" and df["status"] != "ПРОСРОЧЕНО":
    return 0.05 * df["sum"] if df["sum"] > 10000 else 0.03 * df["sum"]
  else:
    return 0
```

```python
df_may_remainder['bonus'] = df_may_remainder.apply(calculate_bonus, axis=1)
df_may_remainder.head()
```

Посчитаем бонусы для каждого менеджера

```python
df_remainder_grouped = df_may_remainder.groupby("sale")["bonus"].sum().reset_index()
df_remainder_grouped
```

Не все менеджеры могли получить остаток, поэтому создадим отдельнуый столбец со всеми менеджерами и соединим их с таблицей с бонусами через left join

```python
managers = pd.Series(df["sale"].unique(), name="sale")[1:] # уберем внутреннюю транзакцию
result = pd.merge(managers, df_remainder_grouped, how="left", on="sale").fillna(0)
result
```
